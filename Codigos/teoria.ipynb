# Reglas para el modelo de mercado con autómatas celulares

## Definiciones
- **Estados de las celdas:**
  - `0` (tengo Acciones): Intención de comprar acciones.
  - `1` (Vendedor o comprar): Posee acciones y está dispuesto a vender.
  - `2` (No tengo acciones): No participa activamente en el mercado.

## Reglas revisadas

| Estado Inicial     | Condición para Cambio                                                   | Estado Final      |
|--------------------|-------------------------------------------------------------------------|------------------ |
| `0` (comprador)    |  \#(2) > \#(1) &&  \#(0) > \#(1)                                        | `1` (vendedor)    |
| `0` (comprador)    |  Probabilidad de expectativa                                            | `2` (no comprador)|
| `1` (vendedor)     |  \#(0) > \#(2) and \#(0) > \#(1) and A > 0                              | `1` (vendedor)    |
| `2` (no comprador) |  \#(0) > \#(2) and \text{probabilidad de regreso}                       | `0` (comprador)   |
| `2` (no comprador) |  \#(0) < \#(2) or \text{probabilidad de permanencia}                   | `2` (no comprador)|

## Ajustes sugeridos

1. **Restricciones basadas en \( B \) y \( A \):**
   - Los compradores (`0`) deben tener \( B > 0 \).
   - Los vendedores (`1`) deben tener \( A > 0 \).

2. **Incentivos y probabilidades ajustadas al mercado:**
   - Introducir una probabilidad de transición de `0` a `2` si \( B = 0 \) o si no hay acciones disponibles en el mercado (\( \#(1) = 0 \)).
   - Aumentar la probabilidad de que `2` pase a `0` si hay un exceso de acciones disponibles (\( \#(1) \gg \#(0) \)).

3. **Balance del mercado:**
   - El sistema debe converger hacia un equilibrio en el que las proporciones de estados (`0`, `1`, `2`) reflejen las dinámicas de oferta y demanda.

4. **Modelo financiero agregado:**
   - Actualizar \( B \) y \( A \) después de cada transición:
     - Si `0` pasa a `1`, decrementar \( B \) y aumentar \( A \).
     - Si `1` pasa a `2`, disminuir \( A \) y reflejar el capital recibido.

## Notas adicionales

1. **Estados adicionales para coherencia:**
   - Los compradores sin dinero (\( B = 0 \)) deben tener una probabilidad alta de transición a `2`.
   - Los vendedores sin acciones (\( A = 0 \)) no pueden permanecer en el estado `1`.

2. **Evolución del sistema:**
   - Asegurarse de que el sistema alcance un equilibrio dinámico entre oferta (`1`) y demanda (`0`).

3. **Simulación computacional:**
   - Implementar estas reglas en un modelo computacional para analizar el comportamiento del sistema bajo diferentes configuraciones iniciales y parámetros de probabilidad.
